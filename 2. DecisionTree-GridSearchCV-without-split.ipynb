```python
import pandas as pd
dataset = pd.read_csv('insurance_pre.csv')
dataset.head()
```

```python
dataset = pd.get_dummies(dataset, drop_first=True, dtype=int)
dataset.head()
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent = dataset[['charges']]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor

params_grid = {
    'criterion': ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    'max_features': ['None', 'sqrt', 'log2'],
    'splitter': ['best', 'random']
}

grid = GridSearchCV(DecisionTreeRegressor(), params_grid, refit=True, verbose=3, n_jobs=-1)
grid.fit(independent, dependent)
```

```python
print(grid.best_params_)
```

```python
age = int(input('Enter your Age: '))
bmi = float(input('Enter your BMI Index: '))
children = int(input('Enter your children count: '))
sex = int(input('Enter your Sex (0 or 1): '))
smoker = int(input('Are you a Smoker (0 or 1): '))

result = grid.predict([[age, bmi, children, sex, smoker]])
# print(result)
print(f"According to your data, your charges is {format(result)}")
```
