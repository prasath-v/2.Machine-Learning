```python
import pandas as pd
dataset = pd.read_csv('insurance_pre.csv')
dataset.head()
```

```python
dataset = pd.get_dummies(dataset, drop_first=True, dtype=int)
dataset.head()
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent = dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(independent, dependent, test_size=0.30, random_state=0)
X_train.head()
```

```python
### Multiple Regression Algorithm
# from sklearn.linear_model import LinearRegression
# regressor = LinearRegression()

### Support Vector Machine (SVM) Algorithm
# from sklearn.svm import SVR
# regressor = SVR(kernel='precomputed', C=1)

### Decision Tree Algorithm
# from sklearn.tree import DecisionTreeRegressor
# regressor = DecisionTreeRegressor(criterion='poisson', max_features=None, splitter='random')

### Random Forest ALgorithm
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=100, criterion='poisson', random_state=0)

regressor.fit(X_train, Y_train)
```

```python
Y_pred = regressor.predict(X_test)
len(Y_pred)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(Y_test, Y_pred)
r_score
```

```python
# Save the model
import pickle
filename = 'finalised_model_Regression_Assignment.sav'
pickle.dump(regressor, open(filename, 'wb'))
```
