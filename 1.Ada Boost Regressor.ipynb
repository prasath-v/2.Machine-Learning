```python
import pandas as pd
dataset = pd.read_csv('50_Startups.csv')
dataset.head()
```

```python
dataset = pd.get_dummies(dataset, drop_first=True, dtype=int)
dataset.head()
```

```python
dataset.columns
```

```python
independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
dependent = dataset[['Profit']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(independent, dependent, test_size=0.30, random_state=0)
Y_test.head(2)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor

regressor = AdaBoostRegressor(estimator=DecisionTreeRegressor(), n_estimators=50, learning_rate=0.5, random_state=0)
regressor.fit(X_train, Y_train)
```

```python
Y_pred = regressor.predict(X_test)
Y_pred
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(Y_test, Y_pred)
r_score 

# max_depth => empty(not added as parameter) => 0.9390509199582147 ✅
# max_depth => 1 => 0.6659275154404327
# max_depth => 2 => 0.8988847000078309
# max_depth => 3 => 0.9122587549769298
```
